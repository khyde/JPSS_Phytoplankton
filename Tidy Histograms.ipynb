```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data82018 = pd.read_csv ('2018-08.csv')
data112018 = pd.read_csv ('2018-11.csv')
data52019 = pd.read_csv ('2019-05.csv')
data102019 = pd.read_csv ('2019-10.csv')
```

```python
data52019
```

```python
saldata = [data82018.salinity, data112018.salinity, data52019.salinity, data102019.salinity]
```

```python
tempdata = [data82018.temperature, data112018.temperature, data52019.temperature, data102019.temperature]
```

```python
chldata = [data82018.chl_ctd, data112018.chl_ctd, data52019.chl_ctd, data102019.chl_ctd]
```

```python
cdomdata = [data82018.cdom_ctd, data112018.cdom_ctd, data52019.cdom_ctd, data102019.cdom_ctd]
```

```python
for element in saldata:
    plt.hist (element)
```

```python
for element in saldata:
    plt.hist (element[1])
```

```python
for element in saldata:
    plt.hist (element[1, 2, 3, 4])
```

```python
for element in saldata:
    plt.hist (element)
    plt.figure ()
```

```python
# the rwidth is meant to squeeze the bars a little so that the differences are observable
```

```python
for element in saldata:
    plt.hist (element, color = ['b'], rwidth = 0.95)
    plt.xlabel ("Salinity (PGU)")
    plt.legend (["Salinity (PGU)"])
    plt.figure ()
```

```python
for element in tempdata:
    plt.hist (element, color = ['r'], rwidth = 0.95)
    plt.xlabel ("Salinity (PGU)")
    plt.legend (["Salinity (PGU)"])
    plt.figure ()
```

```python
for element in chldata:
    plt.hist ( np.log(element), color = ['g'], rwidth = 0.95)
    plt.xlabel ("Chlorophyll (uG/L)")
    plt.legend (["Chlorophyll (uG/L)"])
    plt.figure ()
```

```python
for element in cdomdata:
    plt.hist ( np.log(element), color = ['c'], rwidth = 0.95)
    plt.xlabel ("Dissolved Organic Matter")
    plt.legend (["Dissolved Organic Matter"])
    plt.figure ()
```
