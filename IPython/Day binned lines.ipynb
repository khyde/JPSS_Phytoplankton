```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import numpy as np
import seaborn as sns
```

```python
data82018 = pd.read_csv ('2018-08.csv')
data112018 = pd.read_csv ('2018-11.csv')
data52019 = pd.read_csv ('2019-05.csv')
data102019 = pd.read_csv ('2019-10.csv')
```

```python
date82018 = data82018.date
temp82018 = data82018.temperature
sal82018 = data82018.salinity
chl82018 = data82018.chl_ctd
cdom82018 = data82018.cdom_ctd

date112018 = data112018.date
temp112018 = data112018.temperature
sal112018 = data112018.salinity
chl112018 = data112018.chl_ctd
cdom112018 = data112018.cdom_ctd

date52019 = data52019.date
temp52019 = data52019.temperature
sal52019 = data52019.salinity
chl52019 = data52019.chl_ctd
cdom52019 = data52019.cdom_ctd

date102019 = data102019.date
temp102019 = data102019.temperature
sal102019 = data102019.salinity
chl102019 = data102019.chl_ctd
cdom102019 = data102019.cdom_ctd
```

```python
data82018.index = pd.to_datetime(data82018["date"])
data112018.index = pd.to_datetime(data112018["date"])
data52019.index = pd.to_datetime(data52019["date"])
data102019.index = pd.to_datetime(data102019["date"])
```

```python
avgtemp818 = data82018["temperature"].rolling("1D").mean()
avgtemp1118 = data112018["temperature"].rolling("1D").mean()
avgtemp519 = data52019["temperature"].rolling("1D").mean()
avgtemp1019 = data102019["temperature"].rolling("1D").mean()

avgsal818 = data82018["salinity"].rolling("1D").mean()
avgsal1118 = data112018["salinity"].rolling("1D").mean()
avgsal519 = data52019["salinity"].rolling("1D").mean()
avgsal1019 = data102019["salinity"].rolling("1D").mean()

avgchl818 = data82018["chl_ctd"].rolling("1D").mean()
avgchl1118 = data112018["chl_ctd"].rolling("1D").mean()
avgchl519 = data52019["chl_ctd"].rolling("1D").mean()
avgchl1019 = data102019["chl_ctd"].rolling("1D").mean()

avgcdom818 = data82018["cdom_ctd"].rolling("1D").mean()
avgcdom1118 = data112018["cdom_ctd"].rolling("1D").mean()
avgcdom519 = data52019["cdom_ctd"].rolling("1D").mean()
avgcdom1019 = data102019["cdom_ctd"].rolling("1D").mean()
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharex=True, sharey=False, figsize=(25,25))
ax1.plot(avgtemp818, color='r')
ax1.legend (["Temperature (C)"], shadow=True, fontsize = '16', loc = 'best')
ax1.set_title('August 2018 Time Series', fontsize = '20')
ax1.grid ()
ax2.plot(avgsal818)
ax2.legend (["Salinity (PSU)"], shadow=True, fontsize = '16', loc = 'best')
ax2.grid ()
ax3.plot(avgchl818, color='g')
ax3.legend (["Chlorophyll (uG/L)"], shadow=True, fontsize = '16', loc = 'best')
ax3.grid ()
ax4.plot(avgcdom818, color='c')
ax4.legend (["CDOM"], shadow=True, fontsize = '16', loc = 'best')
ax4.grid ()
ax1.tick_params(axis='both', labelsize=17)
ax2.tick_params(axis='both', labelsize=17)
ax3.tick_params(axis='both', labelsize=17)
ax4.tick_params(axis='both', labelsize=17)
plt.gcf ().autofmt_xdate()
f.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
```
