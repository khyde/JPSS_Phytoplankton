```python
import netCDF4
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from nco import Nco
nco = Nco()
import sys
```

```python
sst1tst = Dataset('D_20191029-MUR-NESGRID-SST.nc', mode='r')

lons = sst1tst.variables['longitude'][:]
lats = sst1tst.variables['latitude'][:]
sst = sst1tst.variables['sst'][:]

sst1tst.close()

lon_0 = lons.mean()
lat_0 = lats.mean()

plt.figure(figsize=(24,12))
m = Basemap(width=3000000,height=1500000,
            resolution='l',projection='tmerc',\
            lat_ts=40,lat_0=lat_0,lon_0=lon_0)

lon, lat = np.meshgrid(lons, lats)
xi, yi = m(lon, lat)

cs = m.pcolor(xi,yi,np.squeeze(sst))

m.drawparallels(np.arange(-80., 81., 10.), labels=[1,0,0,0], fontsize=10)
m.drawmeridians(np.arange(-180., 181., 10.), labels=[0,0,0,1], fontsize=10)

m.drawcoastlines()
m.drawstates()
m.drawcountries()

cbar = m.colorbar(cs, location='bottom', pad="10%")
cbar.set_label(sst)

plt.title('Surface Temperature')

plt.show()
```
