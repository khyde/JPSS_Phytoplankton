```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import numpy as np
import seaborn as sns
```

```python
stations818 = pd.read_csv ('ECOMON stations 2018-08.csv')
stations1118 = pd.read_csv ('ECOMON stations 2018-11.csv')
```

```python
#f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
#ax1.plot(x, y)
#ax1.set_title('Sharing Y axis')
#ax2.scatter(x, y)
```

```python
depth818 = stations818["depth"]
temp818 = stations818["temperature"]
sal818 = stations818["salinity"]
si818 = stations818["si_oh4"]
nit818 = stations818["no2_no3"]
chl818 = stations818["chl_a"]
totchl818 = stations818["tot_chl_a"]

depth1118 = stations1118["depth"]
temp1118 = stations1118["temperature"]
sal1118 = stations1118["salinity"]
si1118 = stations1118["si_oh4"]
nit1118 = stations1118["no2_no3"]
chl1118 = stations1118["chl_a"]
totchl1118 = stations1118["tot_chl_a"]
```

```python
#nit = nitrite/nitrate
#fl = () fl818 = stations818[""]
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, sharey=True, figsize=(25,25))
ax1.set_title('August 2018 Depth Series', fontsize = '20')
ax1.plot(temp818, depth818, color='r')
ax1.plot(sal818, depth818, color ='b')
ax1.legend (["Temperature (°C)", "Salinity (PSU)"], shadow=True, fontsize = '16', loc = 'best')
ax1.tick_params(axis='both', labelsize=17)
ax1.grid ()
ax2.plot(chl818, depth818, 'g')
ax2.plot(totchl818, depth818, '-g')
ax2.legend (["Chlorophyll (uG/L)", "Total Chlorophyll (uG/L)"], shadow=True, fontsize = '16', loc = 'best')
ax2.tick_params(axis='both', labelsize=17)
ax2.grid ()
ax3.plot(nit818, depth818,'m')
ax3.plot(si818, depth818,'c')
ax3.legend (["Nitrite/Nitrate", "Silicate"], shadow=True, fontsize = '16', loc = 'best')
ax3.tick_params(axis='both', labelsize=17)
ax3.grid ()
f.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
```
