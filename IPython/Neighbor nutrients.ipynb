```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import numpy as np
import seaborn as sns

stations818 = pd.read_csv ('ECOMON stations 2018-08.csv')

loop818 = stations818.set_index('discrete_station_id', drop=False)

stations = [129, 133, 135, 136, 137, 138, 139, 140, 141, 142, 143, 144, 145, 146, 147, 148, 149, 150, 151, 152, 153, 154, 155, 156, 157, 158, 159]

loop818["date"] = pd.to_datetime (loop818["date"])

for station in stations:
    try:
        #data = loop818.loc[station,"temperature"]
        f, axs = plt.subplots(4,1, sharey ='all', figsize=(15,10))
        plt.suptitle("Station: {} {}".format(station, (loop818.loc[station, "date"].min().date()))) 
        #plot 1
        plt.subplot(131)
        s1mask = ~np.isnan(loop818.loc[station, "temperature"]) # may need to add a second mask for salinity
        plt.plot(loop818.loc[station,"temperature"][s1mask], loop818.loc[station,"depth"][s1mask], linestyle='-', marker='.', color='r')
        plt.plot(loop818.loc[station,"salinity"][s1mask], loop818.loc[station,"depth"][s1mask], linestyle='-', marker='.', color='b')
        plt.ylim (0,50)
        plt.legend (["Temperature (C)", "Salinity (PSU)"], shadow=True, fontsize = '10', loc = 'best')
        plt.tick_params(labelsize=17)
        plt.gca ().invert_yaxis ()
        plt.grid ()
        #plot 2
        plt.subplot(132)
        s1mask = ~np.isnan(loop818.loc[station, "chl_a"])
        plt.plot(loop818.loc[station,"chl_a"][s1mask], loop818.loc[station,"depth"][s1mask], linestyle='--', marker='X', color='g')
        plt.plot(loop818.loc[station,"tot_chl_a"][s1mask], loop818.loc[station,"depth"][s1mask], linestyle='-', marker='o', color='g')
        plt.ylim (0,50)
        plt.legend (["Chlorophyll (uG/L)", "Total Chlorophyll (uG/L)"], shadow=True, fontsize = '10', loc = 'best')
        plt.tick_params(labelsize=17)
        plt.gca ().invert_yaxis ()
        plt.grid ()
        #plot 3
        plt.subplot(133)
        s1mask = ~np.isnan(loop818.loc[station, "no2_no3"])
        plt.plot(loop818.loc[station,"no2_no3"][s1mask], loop818.loc[station,"depth"][s1mask],linestyle='-', marker='o', color='m')
        plt.plot(loop818.loc[station,"si_oh4"][s1mask], loop818.loc[station,"depth"][s1mask],linestyle='-', marker='o', color='c')
        plt.ylim (0,50)
        plt.legend (["Nitrite/Nitrate", "Silicate"], shadow=True, fontsize = '10', loc = 'best')
        plt.tick_params(labelsize=17)
        plt.gca ().invert_yaxis ()
        plt.grid ()
        plt.savefig("{}.png".format(station))
    except KeyError:
        print ("key does not exist "+str(station))
```
